## Import Libraries

```python
# import libraries

# data analysis
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import plotly.express as px

# download files to our computer
from google.colab import files 
```

## Import Data

```python
# Import and save csv of snapchat political ads data
snap_data = "https://raw.githubusercontent.com/ireneliu0106/Baltimore-city-employee-cluster-analysis/main/PoliticalAdsCombined.csv"
snap2020 = "https://raw.githubusercontent.com/ireneliu0106/Snap_PoliticalAds2020_Regression_Analyses/main/PoliticalAds2020.csv"

#define our initial dataframe
df_snap = pd.read_csv(snap_data)
df_snap2020 = pd.read_csv(snap2020)
```

## Process Data

```python
df_snap.head()
```

```python
# only include data in USD
df_snap_usa = df_snap[df_snap["Currency Code"] == "USD"]
df_snap2020_usa = df_snap2020[df_snap2020["Currency Code"]=="USD"]
```

```python
# new data frame with split value of start date
new = df_snap_usa["StartDate"].str.split("/", n = 1, expand = True)
  
# make separate year column from new data frame
df_snap_usa["Year"]= new[0]
  
# make separate date column from new data frame
df_snap_usa["Date"]= new[1]
  
```

```python
# group the sum of impressions and spending based on year 
df_snap_agg = df_snap_usa.groupby("Year").agg({"Impressions": "sum",
                                                     "Spend": "sum"}).reset_index()
                                                                                                  
```

```python
# look at the sum of impressions and spending in different years
df_snap_agg
```

```python
# create a trendline of number of impressions from 2018 to 2021
year_impression_plot = px.line(df_snap_agg, x = "Year", y = "Impressions",
                            title = "The Number of Impressions of Snapchat Ads from 2018 to 2020",
                            labels = {"Impressions": "Number of Impressions"})

# show the plot
year_impression_plot.show()
# create a scatter plot of impressions vs spending and trendline 
impression_spending_plot = px.scatter(df_snap2020_usa, x="Spend", y="Impressions",trendline="ols",
                                       title = "Impression of Ad as a Function of Spending",
                            labels = {"Spend": "Spending (USD)", "Impressions": "Number of Impressions"})
impression_spending_plot.show() 
```

```python
# save the plots as html files
year_impression_plot.write_html("year_impression_plot.html")
impression_spending_plot.write_html("plotly_impression_spending.html")
# download from google colab
files.download("year_impression_plot.html")
files.download("plotly_impression_spending.html")
```
